```python
# Calculate annual attrition rate
def calculate_attrition_rate(df):
    total_employees = len(df)
    attrition_count = df[df['Attrition'] == 'Yes']['Attrition'].count()
    attrition_rate = (attrition_count / total_employees) * 100
    return attrition_rate

# Group data by distance bins (e.g., 5-mile intervals)
bins = [0, 5, 10, 15, 20, 25, 30]
labels = [f"{i}-{i+5}" for i in bins[:-1]]
df['DistanceBin'] = pd.cut(df['DistanceFromHome'], bins=bins, labels=labels)

# Calculate attrition rate for each distance bin
attrition_by_distance = df.groupby('DistanceBin').apply(calculate_attrition_rate)

# Sort the data by distance
attrition_by_distance = attrition_by_distance.sort_values(ascending=False)

# Create a bar chart with Seaborn, using a transition color from light gray to dark green
color_palette = sns.color_palette("mako", len(attrition_by_distance))
plt.figure(figsize=(10, 6))
sns.barplot(x=attrition_by_distance.index, y=attrition_by_distance.values, palette=color_palette)
plt.xlabel('Distance from Home (km)')  # Updated x-axis label
plt.ylabel('Attrition Rate (%)')
plt.title('Attrition Rate by Distance from Home')
plt.xticks(rotation=0)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

nsmutes.

- **Highest Attrition Rate: Occurs in the 10-15 km distance category (around 20%)**.- **Lower Rates: Employees living very close (0-5 km) or farther away (25-30 km) have attrition rates around 10%**
- **Policies like flexible work arrangements or remote work options could help reduce attrition for employees with moderate commutes**.

```python
# Group by YearsWithCurrentManager and calculate average PerformanceRating
average_rating = df.groupby('YearsWithCurrManager')['PerformanceRating'].mean()

# Calculate overall average performance rating
overall_average = df['PerformanceRating'].mean()

# Line plot with average performance rating per YearsWithCurrentManager
plt.figure(figsize=(8, 6))
plt.plot(average_rating.index, average_rating.values, marker='o', linestyle='-', label='Average Rating by Group')

# Calculate trendline coefficients
m, b = np.polyfit(average_rating.index, average_rating.values, 1)  # Linear trendline (degree 1)

# Generate x values for trendline
x_fit = np.linspace(average_rating.index.min(), average_rating.index.max())

# Calculate y values for trendline
y_fit = m * x_fit + b

# Plot trendline
plt.plot(x_fit, y_fit, color='red', linestyle='--', label='Trendline')


# Add labels and title
plt.xlabel('Years With Current Manager')
plt.ylabel('Performance Rating')
plt.title('Relation Between Years With Current Manager and Performance Rating ')
plt.grid(True)
plt.legend()  # Include legend for all lines

plt.tight_layout()
plt.show()
```

- **Reverse relationship between Year with current manager and Performance rate**.
-  **Very Important insights that clarify that increase of Years with the same manager lead to decrease in Performance rate**.
